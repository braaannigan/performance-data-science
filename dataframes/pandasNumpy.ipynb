```python
import numpy as np
import pandas as pd
```

```python
N = 10000
df = pd.DataFrame(np.random.standard_normal((N,N)))
df.head()
```

```python
def squarePandas(df):
    return df**2
def squareNumpy(df):
    return pd.DataFrame(df.values**2,index=df.index,columns=df.columns)
```

```python
%timeit -n 1 -r 1 squarePandas(df=df)
```

```python
%timeit -n 1 -r 1 squareNumpy(df=df)
```
